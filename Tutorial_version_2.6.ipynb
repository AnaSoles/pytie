# PyTie
#### *Seismic-well tie in Python*
------------
This package reads .LAS files and returns at the end a time-depth table. By calculating your time-depth table with PyTie you can be in control of each step necessary to create the correlation between seismic and well data. The format of files created here are compatible with the free software OpendTect.

## PyTie function: rhob_fill_gardner

This function is based on the Gardner's equation as shown below:

$ \rho = \alpha V_p^{\beta}$  

where, $\rho$ is the density and its unit is in $g/cc$, $V_p$ is the velocity of the P wave in ft/s, $\alpha = 0.23$, $beta = 0.25$ - empirically derived constants that depend on the geology. 

```python
#parameters values - they can change with geology. Standard values: alfa = 0.23 and beta = 0.25:
alfa = 0.25
beta = 0.25
```

```python
RHOB_fill = pytie.rhob_fill_gardner(RHOB, DT, alfa, beta)
```

```python
plt.subplot(1, 2, 1)
plt.plot(RHOB,-DEPTH)
plt.ylim((-1792,-1776))

plt.subplot(1, 2, 2)
plt.plot(RHOB_fill,-DEPTH)
plt.ylim((-1792,-1776))

plt.tight_layout()
```

## PyTie function: DT_fill_gardner

This function is based on the inverse Gardner's equation, as shown below.

$ V_p = \gamma \rho ^4$  

where, $V_p$ is in ft/s, $\gamma$ is an experimental parameter that changes with geology (standard value = 357) and $\rho$ in $g/cc$.

```python
gamma = 357
```

```python
DT_fill = pytie.DT_fill_gardner(RHOB,DT, gamma)
```

```python
plt.subplot(1, 2, 1)
plt.plot(DT,-DEPTH)
plt.ylim((-125,-101))

plt.subplot(1, 2, 2)
plt.plot(DT_fill,-DEPTH)
plt.ylim((-125,-101))

plt.tight_layout()
```

## PyTie function: impedande; Calculating the acoustic impedance

The acoustic impedance is the product of density and velocity of each sample.

$ I = \rho * V$ 

Where, $\rho =  density $ and $V = velocity.$ 

Even though the unity here is not important in terms of time-depth conversion, the shape of the seimogram changes if the DT log is used instead of the velocity of the P wave. This happens because the DT log is a measure of vagarosity of the P wave, that has an 
inverse relationship with the velocity of the P wave.

```python
# Pytie function: calculate_vp; calculates the velocity of the P wave based on the DT log.
```

```python
Vp = pytie.calculate_vp(DT)
Vp
```

```python
I = pytie.impedance(Vp,RHOB)
```

## PyTie function: r_coef; Calculating the reflection coefficients

The reflexion coefficients are an adimensional parameter. Because of this it's not necessary to be worried about units. The coefficients can be found by the following formula:

$R_{coef} = \dfrac{I_2 - I_1}{I_2 + I_1}$

Where: $I$ = Acoustic Impedance

```python
R_coef = pytie.r_coef(I)
```

# Image Cosmetics

The functions bellow don't impact in the calculus of the time-depth table. Because of it they are called image cosmetic functions. Their goal is to improve the visualization of seismic reflectors. Here they are used to make the synthetic seismogram more similar to the seismic section.

## Adding gain to compensate the absorption phenomenon

Stipulate the initial depth of the gain function:

```python
initial_depth = 100
```

```python
gain, seismogram_gain = pytie.gain(initial_depth, DEPTH, R_coef, length, dt, f)
```

```python
title = 'Comparison between seismograms with and without gain'
```

```python
pytie.combo_plot2(top_depth, bottom_depth, depths_markers, markers, RHOB, DEPTH, DT, R_coef, seismogram, seismogram_gain, title)
```

## Loss of frequency with depth due to attenuation

```python
seism_freq = pytie.freq_corr(DEPTH, gain, f, length, dt)
```

```python
title = 'Comparison between the original seismogram and the application of image cosmetics functions'
```

```python
pytie.combo_plot3(top_depth,bottom_depth, depths_markers, markers, RHOB, DEPTH, DT, R_coef, seismogram, seismogram_gain, seism_freq, title)
```
